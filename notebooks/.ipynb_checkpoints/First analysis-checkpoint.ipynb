# Load data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
from sklearn.covariance import EllipticEnvelope
from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM
```

# Load data

```python
df = pd.read_csv("../data/tsfresh_data_cleaned.csv")
df.drop(columns={"Unnamed: 0"}, inplace=True)
df_full = pd.read_csv("../data/dataset.csv")
outliers_predictions = pd.DataFrame()
outliers_predictions.index = df["index"]
```

## Check correaltion to remove variables

```python
corr = df.corr()
sns.heatmap(corr)
plt.show()
```

```python
for col in corr.columns:
    if len(corr[corr[col] == 1]) > 1:
        temp = corr[corr[col] == 1]
        to_be_removed = temp.index.tolist()
        to_be_removed.remove(to_be_removed[0])
        try:
            df.drop(columns=to_be_removed, inplace=True)
        except:
            pass
```

# Analysis

```python
df.set_index("index", drop=True, inplace=True)

df_cycle_1 = df[df["cycle"] == 1]
outliers_cycle_1 = pd.DataFrame()
outliers_cycle_1.index = df_cycle_1.index

df_cycle_2 = df[df["cycle"] == 2]
outliers_cycle_2 = pd.DataFrame()
outliers_cycle_2.index = df_cycle_2.index

df_cycle_3 = df[df["cycle"] == 3]
outliers_cycle_3 = pd.DataFrame()
outliers_cycle_3.index = df_cycle_3.index

df_cycle_4 = df[df["cycle"] == 4]
outliers_cycle_4 = pd.DataFrame()
outliers_cycle_4.index = df_cycle_4.index
```

## Analysis of each time stpe

### Detect anomaly among all time step

```python
iforest = IsolationForest(
    n_estimators=100,
    max_samples="auto",
    contamination=0.05,
    max_features=4,
    bootstrap=False,
    n_jobs=-1,
    random_state=1,
)
pred = iforest.fit_predict(df_full)
df_full["scores"] = iforest.decision_function(df_full)
df_full["anomaly_label"] = pred
```

```python
df_full["anomaly"] = df_full["anomaly_label"].apply(
    lambda x: "outlier" if x == -1 else "inlier"
)
```

```python
fig = sns.histplot(data=df_full, x="scores", hue="anomaly")
plt.show()
```

### Average score and number of outlier per window

```python
df_full["anomaly_binary"] = df_full["anomaly_label"].apply(
    lambda x: 1 if x == -1 else 0
)
```

```python
df_temp = (
    df_full[["day_cycle_window", "anomaly_binary", "scores"]]
    .groupby("day_cycle_window")
    .mean()[["anomaly_binary", "scores"]]
)
```

```python
df = df.merge(df_temp, how="left", left_on="index", right_index=True)
```

## MAD among the classes of window

### Cycle 1

```python
def MAD(col):
    med = col.median()
    mad = (col - med).abs().median()

    threshold1 = med - 3 * mad
    threshold2 = med + 3 * mad
    anomalies = (col - med).abs() > 3 * mad
    return anomalies.copy()
```

```python
outliers_cycle_1 = df_cycle_1.apply(MAD)
outliers_cycle_1 = outliers_cycle_1.add_prefix("MAD_")
```

```python
outliers_cycle_1 /= 200
```

## Elliptic enveloppe

### Cycle 1

```python
cov = EllipticEnvelope(random_state=0, contamination=0.15).fit(df_cycle_1)
```

```python
outliers_cycle_1["Elliptic_envelop"] = cov.predict(df_cycle_1)
```

## Isolation Forest

### Cycle 1

```python
clf = IsolationForest(n_estimators=100, contamination=0.15)
clf.fit(df_cycle_1)
```

```python
outliers_cycle_1["Isolation_forest"] = clf.predict(df_cycle_1)
```

## LOF (Local Outlier factor)

```python
clf = LocalOutlierFactor(n_neighbors=5, contamination=0.15, novelty=True)
clf.fit(df_cycle_1)
```

```python
outliers_cycle_1["LOF"] = clf.predict(df_cycle_1)
```

## OneClass SVM

```python
clf = OneClassSVM(nu=0.15, kernel="rbf", gamma=0.3)  # nu corresponds to contamination
clf.fit(df_cycle_1)
```

```python
outliers_cycle_1["One_Classe_SVM"] = clf.predict(df_cycle_1)
```

# Conlusion part

```python
outliers_cycle_1 = outliers_cycle_1.astype(float)
outliers_cycle_1.replace(-1, 0, inplace=True)
```

```python
outliers_cycle_1.sum()
```

```python
final = pd.DataFrame(outliers_cycle_1.sum(axis=1))
```

```python
plt.hist(final)
plt.show()
```

```python
med = final.median()
mad = (final - med).abs().median()

threshold1 = med - 3 * mad
threshold2 = med + 3 * mad
anomalies = final[(final - med).abs() > 3 * mad].copy()
```

```python
anomalies.sum()
```

```python
def plotAnomalies1D(s, anomalies, threshold1, threshold2):
    """
    s: Pandas Series containing all the points to plot
    anomalies: Pandas Series containing all the points which are anomalies
    threshold1: Float value - minimum threshold to be normal
    threshold2: Float value - maximum threshold to be normal
    """
    plt.figure()
    plt.plot(s, [0] * len(s), "bo")
    plt.plot(anomalies, [0] * len(anomalies), "ro")
    plt.plot([threshold1] * 2, [-1, 1], "g--")
    plt.plot([threshold2] * 2, [-1, 1], "g--")
```

```python
plotAnomalies1D(final.values, anomalies.values, threshold1, threshold2)
```
